# Find out sum of Profit by Subcategory

```python
orders.columns
```

```python
e = (
    orders.groupby(by = "Sub-Category")
    .agg({"Profit": "sum"})
    .sort_values(by = "Profit", ascending=False)
    .round(2)
)
```

```python
e
```

```python
e.plot(kind="bar")
```

```python
e.plot(kind="barh")
```

# Find out average profit by Region and Segement

```python
f = (
    orders.groupby(by = ["Region", "Segment"])
    .agg({"Profit": "mean"})
    .round(2)
)
```

```python
f
```

```python
f.plot(kind="barh")
```

# Show the sum of Sales by Quarter

```python
orders["order quarter"] = orders["Order Date"].dt.to_period("Q")
orders
```

```python
q = (
    orders.groupby(by = "order quarter")
    .agg({"Sales": "sum"})
    .round(2)
)
```

```python
q
```

```python
q.plot(kind="line")
```

# Show monthly sum of quantity

```python
orders["order month"] = orders["Order Date"].dt.to_period("M")
orders
```

```python
m = (
    orders.groupby(by = "order month")
    .agg({"Quantity": "sum"})
    .round(2)
)
```

```python
m
```

```python
m.plot(kind="line")
```

# Show the top 5 States by Sum of Profit

```python
s = (
    orders.groupby(by = "State")
    .agg({"Profit": "sum"})
    .nlargest(5, "Profit")
    .round(1)
)
```

```python
s
```

```python
s.plot(kind="bar")
```
