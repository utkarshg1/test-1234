### If i want natural numbers from 1 to n range(1, n+1)

```python
# Take input from user
n = int(input("Please enter n : "))

# Apply for loop
for i in range(1, n+1):
    print(i**2)
```

### Find out sum of first n natural number
1+2+3+4+5+6+...+n
1. Always intialize sum to 0

```python
list(range(1,11))
```

```python
# Take n as input from user
n = int(input("Please enter natural number : "))

# Initialize sum to 0
s = 0

# Apply for loop
for i in range(1, n+1):
    print(f'Current Number : {i}, Current sum : {s}')
    s = s + i
    print(f'Updated Sum : {s}')
    print('\n=================================================\n')
```

### Write a code to generate product of first n natural numbers

```python
n = int(input("Please enter a natural number :"))

# Intialize product 1
p = 1

# Apply for loop
for i in range(1, n+1):
    p = p*i

# Display the final results
print(f'Product of first {n} natural numbers is : {p}')    
```

### Sum only first 200 elements from a list

```python
s = 0 
for i in d[:200]:
    s = s + i
    
print(f'Sum of first 200 elements from list is {s}')
```

```python
d.index(200)
```

```python
s = 0
for i in d:
    s = s + i
    if d.index(i)==199:
        break

print(f'Sum of first 200 elements of list is {s}')
```

### How to work multiple values in a list

```python
students = [('Utkarsh', 86), ('Raman', 95), ('Aditi', 98), ('Rahul', 78), ('Priya', 57)]
```

```python
students
```

```python
len(students)
```

```python
for i in students:
    print(i)
```

```python
for name, marks in students:
    print(f'Student name : {name}, Marks : {marks}')
```

```python
emp = [(101, "Sarthak", "Engg."), (102, "Aditi", "HR"), (103, "Raman", "Sales")]
len(emp)
```

```python
for i in emp:
    print(i)
```

```python
for idx, name, dept in emp:
    print(f'Employee id : {idx}, Name : {name}, Dept. : {dept}')
```
